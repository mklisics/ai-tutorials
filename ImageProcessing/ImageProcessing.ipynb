```python
# Functions to display images (non OpenCV)
from matplotlib import pyplot as plt
def imshow(image, title=None, size=4):
    img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    fig = plt.figure(figsize=(size,size))
    plt.imshow(img_rgb, figure=fig)
    if title:
        plt.title(title)
    #plt.xticks([])
    #plt.yticks([])
    plt.axis("off")
    plt.show()
    
def imshow_hstack(images, size=4):
    figsize = size*len(images)
    figure, axis = plt.subplots(1,len(images),figsize=(figsize,size),squeeze=False, frameon=True)
    for i in range(len(images)):
        title,image = images[i]
        img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        axis[0,i].imshow(img_rgb)
        axis[0,i].set_axis_off()
        if title:
            axis[0,i].title.set_text(title)
    plt.show()
    
def imshow_info(image, image_name):
    h,w,d = image.shape if len(image.shape) == 3 else (image.shape[0],image.shape[1],1)
    print("Info about", image_name)
    print("    width:", w)
    print("   height:", h)
    print(" channels:", d)
    print("    dtype:", image.dtype)
```

## Image Scaling

```python
# Scaling
imshow_info(img2, "lion.jpg")
```

```python
img2_scaled_up = cv2.resize(img2, None, fx=2, fy=2)
imshow_info(img2_scaled_up, "lion.jpg [scaled up]")
```

```python
img2_scaled_down = cv2.resize(img2, None, fx=0.5, fy=0.5)
imshow_info(img2_scaled_down, "lion.jpg [scaled down]")
```

```python
# Resizing to specific width and height
newx = 64
newy = 64
img2_resized_lin = cv2.resize(img2, (newx, newy), interpolation=cv2.INTER_LINEAR)
imshow_info(img2_resized_lin, "lion.jpg [resized]")
```

```python
img2_resized_near = cv2.resize(img2, (newx, newy), interpolation=cv2.INTER_NEAREST)
img2_resized_area = cv2.resize(img2, (newx, newy), interpolation=cv2.INTER_AREA)
img2_resized_cubic = cv2.resize(img2, (newx, newy), interpolation=cv2.INTER_CUBIC)

imshow_hstack([
    ("Linear", img2_resized_lin),
    ("Nearest", img2_resized_near),
    ("Area", img2_resized_area),
    ("Cubic", img2_resized_cubic)])
```
